```python
import gensim
# Download pretrained weights from https://code.google.com/archive/p/word2vec/
model = gensim.models.KeyedVectors.load_word2vec_format('./GoogleNews-vectors-negative300.bin.gz', binary=True)

```

In corpora, the corpus is never loaded to memory, all corpuses wrapped to special class ``Dataset`` and provide ``__iter__`` method




```python
import pandas as pd
data = pd.read_csv("Aggregated.csv")
words = set(data["prime"].unique()).union(set(data["target"].unique()))
```

```python
diff = words.difference(model.vocab)
```

```python
print(diff)
```

```python
import nltk
nltk.download()
from nltk.corpus import wordnet as wn

```

```python
wn_missing = set()
for word in words:
    if not wn.synsets(word):
        wn_missing.add(word)
```

```python
wn_missing
```

```python
total_missing = wn_missing.union(diff)
```

```python
total_missing
```

```python
cleaned_df = data[~(data["prime"].isin(total_missing) | data["target"].isin(total_missing))]
```

```python
cleaned_df
```

```python
cleaned_df.to_csv("aggregated_clean.csv")
```
